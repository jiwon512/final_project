```python
from konlpy.tag import Okt
from soykeyword.proportion import CorpusbasedKeywordExtractor
import re

okt = Okt()

def clean_text(text):
    if not isinstance(text, str):  # NaN 방지
        return ""
    text = re.sub(r"<[^>]+>", " ", text)   # HTML 태그 제거
    text = re.sub(r"http\S+", " ", text)   # URL 제거
    text = re.sub(r"[^가-힣a-zA-Z0-9\s]", " ", text)  # 특수문자 제거
    text = re.sub(r"\s+", " ", text)       # 공백 정리
    return text.strip()

def tokenize(text):
    text = clean_text(text)
    return [word for word, pos in okt.pos(text, stem=True) if pos in ["Noun", "Alpha"]]

# extractor 세팅
corpusbased_extractor = CorpusbasedKeywordExtractor(
    min_tf=20,    # 최소 등장빈도 (너무 낮게 하면 잡음 많음)
    min_df=2,    # 최소 문서빈도
    tokenize=tokenize,
    verbose=True
)

# 전처리 적용한 docs로 학습
docs_cleaned = [clean_text(doc) for doc in docs]
corpusbased_extractor.train(docs_cleaned)
```

```python
def keyword_docs(keyword):
    documents = corpusbased_extractor.get_document_index(keyword)
    keywords = corpusbased_extractor.extract_from_docs(
        documents,
        min_score=0.8,
        min_frequency=100
    )

    return keywords[:30]
```
