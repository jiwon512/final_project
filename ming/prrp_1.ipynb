```python
import pandas as pd

# click_date → datetime 변환
ads_info["click_date"] = pd.to_datetime(ads_info["click_date"])

# 날짜만 추출해서 그룹화
daily_clicks = ads_info.groupby(ads_info["click_date"].dt.date)["click_key"].count()

# 클릭 수가 가장 많은 날짜 찾기
max_date = daily_clicks.idxmax()
max_clicks = daily_clicks.max()

print("📌 클릭 수가 가장 많은 날짜:", max_date)
print("📌 클릭 수:", max_clicks)

# 상위 10일 보기
print("\n📌 클릭 수 TOP 10 일자")
print(daily_clicks.sort_values(ascending=False).head(10))
```

```python
import pandas as pd

# 1) 타입 보정
ads_info["click_date"] = pd.to_datetime(ads_info["click_date"], errors="coerce")

# 2) 날짜만 분리 (초/분 제거)
ads_info["date"] = ads_info["click_date"].dt.date

# 3) click_time이 0~23인지 점검(선택)
ads_info = ads_info[ads_info["click_time"].between(0, 23)]

# 4) 날짜 × 시간대별 클릭 수 집계
daily_hourly = (
    ads_info.groupby(["date", "click_time"])
            .size()
            .reset_index(name="clicks")
)

# 5) 날짜별로 '클릭 수 최대' 시간대만 추출
idx = daily_hourly.groupby("date")["clicks"].idxmax()
peak_hours = daily_hourly.loc[idx].sort_values("date")

print("📌 날짜별 최다 클릭 시간대 (정상 계산)")
print(peak_hours.head(10))
```

```python
overall_top10 = daily_hourly.sort_values("clicks", ascending=False).head(10)
print(overall_top10)
```

```python
# 시간대별(0~23) 전체 클릭 수 합계
hourly_clicks = daily_hourly.groupby("click_time")["clicks"].sum().reset_index()

# 클릭 수 기준 내림차순 정렬 후 TOP 10
hourly_top10 = hourly_clicks.sort_values("clicks", ascending=False).head(10)

print("📌 전체 시간대별 클릭 수 TOP 10")
print(hourly_top10)
```
