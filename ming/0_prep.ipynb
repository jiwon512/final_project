```python
# 기준 키
keys = ["mda_idx", "dvc_idx", "click_key"]

# 광고적립에서 전환된 키만 추출 (중복 제거)
converted_keys = ads_reward[keys].drop_duplicates()

# 광고참여에 전환여부 컬럼 추가
ads_participation = ads_participation.merge(
    converted_keys.assign(converted=1),
    on=keys,
    how="left"
)

ads_participation["converted"] = ads_participation["converted"].fillna(0).astype(int)
```

```python
keys = ["mda_idx", "dvc_idx", "click_key"]

# 광고적립 키
reward_keys = ads_reward[keys].drop_duplicates()

# 광고참여 키
part_keys = ads_participation[keys].drop_duplicates()

# 광고적립에만 있고 광고참여엔 없는 것
missing_in_part = reward_keys.merge(part_keys, on=keys, how="left", indicator=True)
missing_in_part = missing_in_part[missing_in_part["_merge"] == "left_only"]
```
