```python
# 도메인명 정제

ads_list = pd.read_csv("/Users/Jiwon/Documents/GitHub/final_project/Jiwon/광고도메인.csv")
ads_list

import re
import pandas as pd

ALLOWED = [
    "게임","뷰티","커머스","의료/건강","금융","미디어/컨텐츠","식음료","생활",
    "채용","운세","식당/카페","운동/스포츠","교육","비영리/공공","기타"
]

PATTERNS = [
    (r'\b게임\b', "게임"),
    (r'\b뷰티\b', "뷰티"),
    (r'\b커머스\b', "커머스"),
    (r'의료\s*/?\s*건강', "의료/건강"),
    (r'\b금융\b', "금융"),
    (r'미디어\s*/?\s*[컨콘]텐츠|미디어콘텐츠', "미디어/컨텐츠"),
    (r'\b식음료\b', "식음료"),
    (r'\b생활\b', "생활"),
    (r'\b채용\b', "채용"),
    (r'\b운세\b', "운세"),
    (r'식당\s*/?\s*카페|식당카페', "식당/카페"),
    (r'운동\s*/?\s*스포츠|운동스포츠', "운동/스포츠"),
    (r'\b교육\b', "교육"),
    (r'비영리\s*/?\s*공공', "비영리/공공"),
    (r'\b기타\b', "기타"),
]

def normalize_domain(text: str) -> str:
    if pd.isna(text):
        return "기타"
    s = str(text)

    # 1) JSON 형태면 값만 추출
    m = re.search(r'"domain"\s*:\s*"([^"]+)"', s)
    if m:
        s = m.group(1)

    # 2) 포매팅/개행 제거
    s = re.sub(r'(\\n|[\r\n])+', ' ', s)           # 개행 제거
    s = re.sub(r'[`"*#]+', '', s)                 # 서식문자 제거
    s = s.strip()

    # 3) 허용 라벨 패턴 찾기
    for pat, lab in PATTERNS:
        if re.search(pat, s):
            return lab

    # 4) 라벨명이 정확히 쓰인 경우(여분 텍스트 동반)
    for lab in ALLOWED:
        if lab in s:
            return lab

    return "기타"

# 사용: df["result"] → 표준라벨
ads_list["domain"] = ads_list["domain"].apply(normalize_domain)
```

```python
# 유저 참여 도메인 10개 이상, 5개 10개 사이, 5개 이하, 1개만 참여 구분 분석

# 유저별 참여 도메인 수 계산
user_domain_count = mda_270.groupby('dvc_idx')['domain'].nunique().reset_index()
user_domain_count.columns = ['dvc_idx', 'domain_count']

print(f"총 유저 수: {len(user_domain_count)}")
print(f"도메인 참여 수 분포:")
print(user_domain_count['domain_count'].value_counts().sort_index())

# 도메인 참여 수에 따른 유저 분류
def classify_user_domain_diversity(domain_count):
    if domain_count >= 10:
        return '다양성 탐험가 (10개+)'
    elif domain_count >= 6:
        return '적극적 참여자 (6-9개)'
    elif domain_count >= 2:
        return '선택적 참여자 (2-5개)'
    else:
        return '전문 참여자 (1개)'

user_domain_count['user_type'] = user_domain_count['domain_count'].apply(classify_user_domain_diversity)

# 분류 결과
segment_distribution = user_domain_count['user_type'].value_counts()
print("\n유저 도메인 다양성 세그먼트:")
print(segment_distribution)

# 비율 계산
segment_percentage = (segment_distribution / len(user_domain_count) * 100).round(2)
print("\n세그먼트 비율:")
for segment, percentage in segment_percentage.items():
    print(f"{segment}: {percentage}%")
```

```python
# 각 세그먼트의 상세 행동 분석
def analyze_segments():
    # 유저 타입을 원본 데이터에 매핑
    mda_270_with_segments = mda_270.merge(user_domain_count[['dvc_idx', 'user_type']], on='dvc_idx')
    
    # 세그먼트별 성과 분석
    segment_analysis = mda_270_with_segments.groupby('user_type').agg({
        'dvc_idx': 'nunique',           # 유저 수
        'click_key': 'count',           # 총 클릭수
        'conversion': 'sum',            # 총 전환수
        'domain': 'nunique'             # 참여 도메인 수 (검증용)
    }).round(2)
    
    # 유저당 평균 지표 계산
    segment_analysis['avg_clicks_per_user'] = segment_analysis['click_key'] / segment_analysis['dvc_idx']
    segment_analysis['avg_conversions_per_user'] = segment_analysis['conversion'] / segment_analysis['dvc_idx']
    segment_analysis['conversion_rate'] = (segment_analysis['conversion'] / segment_analysis['click_key'] * 100).round(2)
    
    return mda_270_with_segments, segment_analysis

mda_270_segments, segment_performance = analyze_segments()

print("\n세그먼트별 성과 분석:")
print(segment_performance)
```

### 매체 270 유저 중심 테이블 만들기

```python
import pandas as pd
import numpy as np
from datetime import datetime

# 기본 유저 피처 분석

def create_user_feature_table():
    # datetime 변환
    mda_270['click_date'] = pd.to_datetime(mda_270['click_date'])
    mda_270['hour'] = mda_270['click_date'].dt.hour
    
    # 기본 집계
    user_basic = mda_270.groupby('dvc_idx').agg({
        'user_ip': 'nunique',                    # IP 개수
        'click_key': 'count',                    # 클릭 수
        'conversion': ['sum', lambda x: x.mean() if len(x) else 0],  # 전환수, 전환율
        'click_day': 'nunique', # 방문 날짜 수, 첫/마지막 방문
        'rwd_idx': 'nunique'                     # 리워드 다양성
    })
    
    # 컬럼명 정리
    user_basic.columns = ['ip_count', 'total_clicks', 'total_conversions', 'avg_conversion_rate',
                         'visit_days', 'reward_diversity']
    
    return user_basic

user_features = create_user_feature_table()
user_features
```

```python
# 시간대 분석

def add_time_features():
    # 주요 시간대 (가장 많이 활동하는 시간대)
    user_main_hour = mda_270.groupby('dvc_idx')['hour'].agg(lambda x: x.mode()[0] if len(x.mode()) > 0 else 0)
    
    # 시간대 다양성 (활동한 시간대 수)
    user_hour_diversity = mda_270.groupby('dvc_idx')['hour'].nunique()
    
    # 주중/주말 패턴
    mda_270['is_weekend'] = mda_270['click_date'].dt.weekday >= 5
    user_weekend_ratio = mda_270.groupby('dvc_idx')['is_weekend'].mean()
    
    time_features = pd.DataFrame({
        'main_active_hour': user_main_hour,
        'hour_diversity': user_hour_diversity,
        'weekend_activity_ratio': user_weekend_ratio
    })
    
    return time_features

time_features = add_time_features()
time_features
```

```python
# 광고 카테고리 타입 피처

def add_ad_category_features():
    # 주요 광고 카테고리/타입 (최빈값)
    user_main_category = mda_270.groupby('dvc_idx')['ads_category'].agg(lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown')
    user_main_type = mda_270.groupby('dvc_idx')['ads_type'].agg(lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown')
    user_main_os = mda_270.groupby('dvc_idx')['ads_os_type'].agg(lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown')
    
    # 카테고리/다양성
    category_diversity = mda_270.groupby('dvc_idx')['ads_category'].nunique()
    
    ad_features = pd.DataFrame({
        'main_ad_category': user_main_category,
        'main_ad_type': user_main_type,
        'main_os_type': user_main_os,
        'ad_category_diversity': category_diversity
    })
    
    return ad_features

ad_features = add_ad_category_features()
ad_features
```

```python
# 리조인 광고 참여 분석

# 이거는 뭔가 전환 된 거중에 rejoinable 이 있으면 재 참여율이 얼마나 되는지 분석 => 필요할까? 흠...
```

```python
# 도메인 관련

def add_domain_composition():
    # 전체 클릭에서 도메인 구성비
    user_domain_clicks = mda_270.groupby(['dvc_idx', 'domain']).size().unstack(fill_value=0)
    user_domain_ratios = user_domain_clicks.div(user_domain_clicks.sum(axis=1), axis=0)
    
    # 주요 도메인
    user_main_domain = user_domain_ratios.idxmax(axis=1)
    user_domain_concentration = user_domain_ratios.max(axis=1)  # 가장 집중한 도메인 비율
    
    # 전환이 발생한 광고의 도메인 구성비
    conversion_data = mda_270[mda_270['conversion'] == 1]
    if len(conversion_data) > 0:
        user_conv_domain_clicks = conversion_data.groupby(['dvc_idx', 'domain']).size().unstack(fill_value=0)
        user_conv_domain_ratios = user_conv_domain_clicks.div(user_conv_domain_clicks.sum(axis=1), axis=0).fillna(0)
        user_main_conv_domain = user_conv_domain_ratios.idxmax(axis=1)
    else:
        user_main_conv_domain = pd.Series(index=mda_270['dvc_idx'].unique(), dtype='object').fillna('None')
    
    domain_features = pd.DataFrame({
        'main_domain': user_main_domain,
        'domain_concentration': user_domain_concentration,
        'domain_diversity': user_domain_clicks.gt(0).sum(axis=1),
        'main_conversion_domain': user_main_conv_domain
    })
    
    # 주요 도메인별 비율 (상위 도메인들만)
    top_domains = user_domain_clicks.sum().nlargest(5).index
    for domain in top_domains:
        if domain in user_domain_ratios.columns:
            domain_features[f'{domain}_ratio'] = user_domain_ratios[domain]
    
    return domain_features, user_domain_ratios

domain_features, domain_ratios = add_domain_composition()
domain_features
```

```python
domain_ratios
```

```python
# 전환 

def add_conversion_features():
    # 전환한 광고의 평균 리워드 단가
    conversion_data = mda_270[mda_270['conversion'] > 0]
    
    if len(conversion_data) > 0:
        user_avg_conversion_reward = conversion_data.groupby('dvc_idx')['media_price'].mean()
        user_conversion_ad_counts = conversion_data.groupby('dvc_idx')['rwd_idx'].nunique()
    else:
        user_avg_conversion_reward = pd.Series(index=mda_270['dvc_idx'].unique(), dtype=float).fillna(0)
        user_conversion_ad_counts = pd.Series(index=mda_270['dvc_idx'].unique(), dtype=int).fillna(0)
    
    conversion_features = pd.DataFrame({
        'avg_conversion_reward': user_avg_conversion_reward,
        'conversion_ad_diversity': user_conversion_ad_counts
    })
    
    return conversion_features

conversion_features = add_conversion_features()
conversion_features
```

```python
def create_final_user_table():
    # 모든 피처 결합
    final_user_table = user_features.copy()
    
    # 각 피처 테이블 병합
    for feature_table in [time_features, ad_features, domain_features, conversion_features]:
        final_user_table = final_user_table.merge(feature_table, left_index=True, right_index=True, how='left')
    
    # 결측치 처리
    final_user_table = final_user_table.fillna(0)
    
    # 파생 변수 추가
    final_user_table['conversion_rate'] = final_user_table['total_conversions'] / final_user_table['total_clicks']
    final_user_table['clicks_per_day'] = final_user_table['total_clicks'] / final_user_table['visit_days']
    
    return final_user_table

final_user_features = create_final_user_table()
final_user_features

print(f"최종 유저 피처 테이블 생성 완료!")
print(f"유저 수: {len(final_user_features)}")
print(f"피처 수: {len(final_user_features.columns)}")
print("\n주요 피처들:")
print(final_user_features.columns.tolist())
```

```python
def find_optimal_clusters(X_scaled, max_k=8):
    inertias = []
    silhouette_scores = []
    K_range = range(2, max_k + 1)
    
    print("최적 클러스터 수 탐색 중...")
    
    for k in K_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(X_scaled)
        
        inertias.append(kmeans.inertia_)
        sil_score = silhouette_score(X_scaled, cluster_labels)
        silhouette_scores.append(sil_score)
        
        print(f"k={k}: Inertia={kmeans.inertia_:.0f}, Silhouette={sil_score:.3f}")
    
    # 시각화
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Elbow Method
    ax1.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
    ax1.set_xlabel('Number of Clusters (k)')
    ax1.set_ylabel('Inertia')
    ax1.set_title('Elbow Method for Optimal k')
    ax1.grid(True, alpha=0.3)
    
    # Silhouette Score
    ax2.plot(K_range, silhouette_scores, 'ro-', linewidth=2, markersize=8)
    ax2.set_xlabel('Number of Clusters (k)')
    ax2.set_ylabel('Silhouette Score')
    ax2.set_title('Silhouette Analysis')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 최적 k 추천
    best_k = K_range[np.argmax(silhouette_scores)]
    print(f"\n추천 클러스터 수: {best_k} (Silhouette Score: {max(silhouette_scores):.3f})")
    
    return best_k, silhouette_scores

# 최적 클러스터 수 찾기
optimal_k, sil_scores = find_optimal_clusters(X_scaled)
```

```python
def run_clustering_fixed(X_scaled, n_clusters):
    print(f"\n{n_clusters}개 클러스터로 클러스터링 실행 중...")
    
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    
    # 결과를 원본 데이터에 추가
    clustering_data['cluster'] = cluster_labels
    
    # 실제 존재하는 컬럼으로만 집계
    available_cols = {
        'dvc_idx': 'count',
        'total_clicks': 'mean',
        'total_conversions': 'mean',
        'conversion_rate': 'mean',
        'domain_diversity': 'mean'
    }
    
    # 추가 컬럼이 있는지 확인하고 포함
    if 'total_revenue' in clustering_data.columns:
        available_cols['total_revenue'] = 'mean'
    if 'max_domain_ratio' in clustering_data.columns:
        available_cols['max_domain_ratio'] = 'mean'
    
    cluster_summary = clustering_data.groupby('cluster').agg(available_cols).round(2)
    
    # 컬럼명 정리
    new_columns = ['user_count', 'avg_clicks', 'avg_conversions', 'avg_conv_rate', 'avg_domain_div']
    if 'total_revenue' in available_cols:
        new_columns.append('avg_revenue')
    if 'max_domain_ratio' in available_cols:
        new_columns.append('avg_specialization')
    
    cluster_summary.columns = new_columns
    
    print("\n클러스터별 기본 특성:")
    print(cluster_summary)
    
    return clustering_data, cluster_summary, kmeans

# 수정된 함수로 클러스터링 실행
clustered_data, cluster_stats, model = run_clustering_fixed(X_scaled, 4)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List, Tuple, Optional, Union
import warnings
warnings.filterwarnings('ignore')

class UserSegmentationPipeline:
    """
    광고 플랫폼 유저 세그먼테이션을 위한 종합 파이프라인 (매체사별 분석 지원)
    
    주요 기능:
    1. 매체사별 데이터 필터링
    2. 유저별 종합 피처 생성
    3. 자동 클러스터링 (최적 k 탐색)
    4. 클러스터 특성 분석 및 시각화
    5. 매체사별/통합 비즈니스 인사이트 생성
    """
    
    def __init__(self, user_id_col: str = 'dvc_idx', 
                 media_id_col: str = 'mda_idx',
                 domain_col: str = 'domain'):
        self.user_id_col = user_id_col
        self.media_id_col = media_id_col
        self.domain_col = domain_col
        
        # 초기화
        self.user_features = None
        self.clustered_data = None
        self.cluster_model = None
        self.scaler = None
        self.label_encoders = {}
        self.cluster_summary = None
        self.optimal_k = None
        
        # 매체사별 분석 결과 저장
        self.media_results = {}
        self.current_media_id = None
        
    def get_available_medias(self, df: pd.DataFrame) -> List[int]:
        """사용 가능한 매체사 목록 반환"""
        media_list = sorted(df[self.media_id_col].unique())
        media_counts = df[self.media_id_col].value_counts().sort_index()
        
        print("=== 사용 가능한 매체사 ===")
        for media_id in media_list:
            count = media_counts[media_id]
            users = df[df[self.media_id_col] == media_id][self.user_id_col].nunique()
            print(f"매체사 {media_id}: {count:,}개 레코드, {users:,}명 유저")
        
        return media_list
    
    def filter_by_media(self, df: pd.DataFrame, media_id: int, 
                       min_users: int = 50, min_records: int = 100) -> pd.DataFrame:
        """
        특정 매체사로 데이터 필터링
        
        Args:
            df: 전체 데이터
            media_id: 매체사 ID
            min_users: 최소 유저 수 (기본값: 50)
            min_records: 최소 레코드 수 (기본값: 100)
        
        Returns:
            filtered_df: 필터링된 데이터
        """
        # 매체사 필터링
        media_df = df[df[self.media_id_col] == media_id].copy()
        
        # 데이터 검증
        if len(media_df) < min_records:
            raise ValueError(f"매체사 {media_id}의 레코드가 {len(media_df)}개로 최소 기준({min_records})에 미달합니다.")
        
        unique_users = media_df[self.user_id_col].nunique()
        if unique_users < min_users:
            raise ValueError(f"매체사 {media_id}의 유저가 {unique_users}명으로 최소 기준({min_users})에 미달합니다.")
        
        print(f"매체사 {media_id} 데이터: {len(media_df):,}개 레코드, {unique_users:,}명 유저")
        self.current_media_id = media_id
        
        return media_df
    
    def create_user_features(self, df: pd.DataFrame) -> pd.DataFrame:
        """
        원본 데이터에서 유저별 종합 피처 생성
        
        Args:
            df: 광고 클릭 데이터 (click_key, ads_idx, dvc_idx, conversion, domain 등 포함)
        
        Returns:
            user_features: 유저별 피처 데이터프레임
        """
        print("유저 피처 생성 중...")
        
        # 시간 관련 피처 생성
        df['click_date'] = pd.to_datetime(df['click_date'])
        df['hour'] = df['click_date'].dt.hour
        df['is_weekend'] = df['click_date'].dt.weekday >= 5
        
        # 1. 기본 활동 피처
        basic_features = df.groupby(self.user_id_col).agg({
            'user_ip': 'nunique',
            'click_key': 'count',
            'conversion': ['sum', 'mean'],
            'contract_price': 'sum',
            'click_date': 'nunique',
            'rwd_idx': 'nunique'
        })
        
        basic_features.columns = ['ip_count', 'total_clicks', 'total_conversions', 
                                'avg_conversion_rate', 'total_revenue', 'visit_days', 'reward_diversity']
        
        # 2. 시간 패턴 피처
        time_features = df.groupby(self.user_id_col).agg({
            'hour': ['mean', lambda x: x.mode()[0] if len(x.mode()) > 0 else 0, 'nunique'],
            'is_weekend': 'mean'
        })
        time_features.columns = ['avg_hour', 'main_active_hour', 'hour_diversity', 'weekend_activity_ratio']
        
        # 3. 다양성 피처
        diversity_features = df.groupby(self.user_id_col).agg({
            'ads_idx': 'nunique',
            self.domain_col: 'nunique',
            'ads_category': 'nunique',
            'ads_type': 'nunique'
        })
        diversity_features.columns = ['ad_diversity', 'domain_diversity', 'category_diversity', 'type_diversity']
        
        # 4. 선호도 피처 (최빈값)
        preference_features = df.groupby(self.user_id_col).agg({
            self.domain_col: lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown',
            'ads_category': lambda x: x.mode()[0] if len(x.mode()) > 0 else 0,
            'ads_type': lambda x: x.mode()[0] if len(x.mode()) > 0 else 0,
            'ads_os_type': lambda x: x.mode()[0] if len(x.mode()) > 0 else 0
        })
        preference_features.columns = ['main_domain', 'main_category', 'main_type', 'main_os']
        
        # 5. 도메인 구성비
        domain_features = self._create_domain_ratios(df)
        
        # 6. 전환 관련 피처
        conversion_features = self._create_conversion_features(df)
        
        # 모든 피처 통합
        user_features = basic_features.copy()
        for feature_table in [time_features, diversity_features, preference_features, 
                             domain_features, conversion_features]:
            user_features = user_features.merge(feature_table, left_index=True, right_index=True, how='left')
        
        # 결측치 처리
        numeric_cols = user_features.select_dtypes(include=[np.number]).columns
        user_features[numeric_cols] = user_features[numeric_cols].fillna(0)
        
        categorical_cols = user_features.select_dtypes(include=['object']).columns
        user_features[categorical_cols] = user_features[categorical_cols].fillna('Unknown')
        
        # 파생 변수
        user_features['conversion_rate'] = user_features['total_conversions'] / user_features['total_clicks']
        user_features['revenue_per_click'] = user_features['total_revenue'] / user_features['total_clicks']
        user_features['clicks_per_day'] = user_features['total_clicks'] / user_features['visit_days']
        
        # 무한값 처리
        user_features = user_features.replace([np.inf, -np.inf], 0)
        
        self.user_features = user_features.reset_index()
        print(f"유저 피처 생성 완료: {self.user_features.shape}")
        
        return self.user_features
    
    def _create_domain_ratios(self, df: pd.DataFrame) -> pd.DataFrame:
        """도메인 참여 비율 계산"""
        user_domain_matrix = df.groupby([self.user_id_col, self.domain_col]).size().unstack(fill_value=0)
        user_domain_ratios = user_domain_matrix.div(user_domain_matrix.sum(axis=1), axis=0)
        
        # 상위 5개 도메인만 선택
        top_domains = user_domain_matrix.sum().nlargest(5).index
        domain_features = pd.DataFrame(index=user_domain_ratios.index)
        
        for domain in top_domains:
            if domain in user_domain_ratios.columns:
                domain_features[f'{domain}_ratio'] = user_domain_ratios[domain]
        
        # 도메인 전문성 지표
        domain_features['max_domain_ratio'] = user_domain_ratios.max(axis=1)
        domain_features['domain_entropy'] = -np.sum(
            user_domain_ratios * np.log(user_domain_ratios + 1e-10), axis=1
        )
        
        return domain_features
    
    def _create_conversion_features(self, df: pd.DataFrame) -> pd.DataFrame:
        """전환 관련 피처 생성"""
        conversion_data = df[df['conversion'] > 0]
        all_users = pd.DataFrame(index=df[self.user_id_col].unique())
        
        if len(conversion_data) > 0:
            conv_features = conversion_data.groupby(self.user_id_col).agg({
                'rwd_idx': 'mean',
                'ads_idx': 'nunique',
                self.domain_col: lambda x: x.mode()[0] if len(x.mode()) > 0 else 'No_Conversion'
            })
            conv_features.columns = ['avg_conversion_reward', 'conversion_ad_diversity', 'main_conversion_domain']
        else:
            conv_features = pd.DataFrame(columns=['avg_conversion_reward', 'conversion_ad_diversity', 'main_conversion_domain'])
        
        conversion_features = all_users.merge(conv_features, left_index=True, right_index=True, how='left')
        conversion_features = conversion_features.fillna({'avg_conversion_reward': 0, 
                                                        'conversion_ad_diversity': 0,
                                                        'main_conversion_domain': 'No_Conversion'})
        
        return conversion_features
    
    def prepare_clustering_features(self, categorical_cols: List[str] = None) -> Tuple[np.ndarray, List[str]]:
        """클러스터링용 피처 준비 및 전처리"""
        if self.user_features is None:
            raise ValueError("먼저 create_user_features()를 실행하세요.")
        
        if categorical_cols is None:
            categorical_cols = ['main_domain', 'main_conversion_domain']
        
        # 범주형 변수 인코딩
        for col in categorical_cols:
            if col in self.user_features.columns:
                le = LabelEncoder()
                self.user_features[f'{col}_encoded'] = le.fit_transform(self.user_features[col].astype(str))
                self.label_encoders[col] = le
        
        # 클러스터링용 피처 선택
        clustering_features = [
            'total_clicks', 'total_conversions', 'conversion_rate', 'domain_diversity',
            'weekend_activity_ratio', 'hour_diversity', 'clicks_per_day'
        ]
        
        # 인코딩된 범주형 피처 추가
        encoded_features = [f'{col}_encoded' for col in categorical_cols if col in self.user_features.columns]
        clustering_features.extend(encoded_features)
        
        # 도메인 비율 피처 추가
        ratio_features = [col for col in self.user_features.columns if '_ratio' in col and col != 'weekend_activity_ratio']
        clustering_features.extend(ratio_features)
        
        # 사용 가능한 피처만 선택
        available_features = [col for col in clustering_features if col in self.user_features.columns]
        
        X = self.user_features[available_features].copy()
        
        # 이상치 제거 (99 percentile 기준)
        for col in ['total_clicks', 'total_revenue']:
            if col in X.columns:
                percentile_99 = np.percentile(X[col], 99)
                X[col] = X[col].clip(upper=percentile_99)
        
        # 무한값 및 NaN 처리
        X = X.replace([np.inf, -np.inf], np.nan).fillna(0)
        
        print(f"클러스터링 피처 준비 완료: {len(available_features)}개 피처")
        
        return X.values, available_features
    
    def find_optimal_clusters(self, X: np.ndarray, max_k: int = 8) -> int:
        """최적 클러스터 수 탐색"""
        print("최적 클러스터 수 탐색 중...")
        
        # 표준화
        self.scaler = StandardScaler()
        X_scaled = self.scaler.fit_transform(X)
        
        inertias = []
        silhouette_scores = []
        K_range = range(2, max_k + 1)
        
        for k in K_range:
            kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
            cluster_labels = kmeans.fit_predict(X_scaled)
            
            inertias.append(kmeans.inertia_)
            silhouette_avg = silhouette_score(X_scaled, cluster_labels)
            silhouette_scores.append(silhouette_avg)
            
            print(f"k={k}: Silhouette Score={silhouette_avg:.3f}")
        
        self.optimal_k = K_range[np.argmax(silhouette_scores)]
        print(f"최적 클러스터 수: {self.optimal_k}")
        
        return self.optimal_k
    
    def perform_clustering(self, X: np.ndarray, n_clusters: int = None) -> pd.DataFrame:
        """클러스터링 수행"""
        if n_clusters is None:
            n_clusters = self.optimal_k
        
        if self.scaler is None:
            self.scaler = StandardScaler()
            X_scaled = self.scaler.fit_transform(X)
        else:
            X_scaled = self.scaler.transform(X)
        
        print(f"{n_clusters}개 클러스터로 클러스터링 수행 중...")
        
        self.cluster_model = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        cluster_labels = self.cluster_model.fit_predict(X_scaled)
        
        # 클러스터 라벨 추가
        self.user_features['cluster'] = cluster_labels
        
        # 클러스터별 특성 분석
        self.cluster_summary = self.user_features.groupby('cluster').agg({
            self.user_id_col: 'count',
            'total_clicks': 'mean',
            'total_conversions': 'mean',
            'conversion_rate': 'mean',
            'domain_diversity': 'mean',
            'max_domain_ratio': 'mean'
        }).round(2)
        
        self.cluster_summary.columns = ['user_count', 'avg_clicks', 'avg_conversions', 
                                      'avg_conv_rate', 'avg_domain_diversity', 'avg_specialization']
        
        self.clustered_data = self.user_features.copy()
        
        print("클러스터링 완료!")
        print(self.cluster_summary)
        
        return self.clustered_data
    
    def analyze_clusters(self) -> Dict:
        """클러스터별 상세 분석"""
        if self.clustered_data is None:
            raise ValueError("먼저 클러스터링을 수행하세요.")
        
        analysis_results = {}
        
        for cluster_id in sorted(self.clustered_data['cluster'].unique()):
            cluster_data = self.clustered_data[self.clustered_data['cluster'] == cluster_id]
            
            # 도메인 선호도
            domain_prefs = {}
            ratio_cols = [col for col in cluster_data.columns if '_ratio' in col and col != 'weekend_activity_ratio']
            for col in ratio_cols:
                avg_ratio = cluster_data[col].mean()
                if avg_ratio > 0.05:  # 5% 이상인 것만
                    domain_prefs[col.replace('_ratio', '')] = avg_ratio
            
            analysis_results[cluster_id] = {
                'size': len(cluster_data),
                'percentage': len(cluster_data) / len(self.clustered_data) * 100,
                'avg_clicks': cluster_data['total_clicks'].mean(),
                'avg_conversions': cluster_data['total_conversions'].mean(),
                'conversion_rate': cluster_data['conversion_rate'].mean(),
                'domain_diversity': cluster_data['domain_diversity'].mean(),
                'specialization': cluster_data['max_domain_ratio'].mean(),
                'weekend_activity': cluster_data['weekend_activity_ratio'].mean(),
                'domain_preferences': domain_prefs,
                'main_domain': cluster_data['main_domain'].mode()[0] if len(cluster_data['main_domain'].mode()) > 0 else 'Mixed'
            }
        
        return analysis_results
    
    def generate_cluster_names(self, analysis_results: Dict) -> Dict[int, str]:
        """클러스터별 비즈니스 네이밍"""
        cluster_names = {}
        
        for cluster_id, info in analysis_results.items():
            avg_clicks = info['avg_clicks']
            conversion_rate = info['conversion_rate']
            specialization = info['specialization']
            main_domain = info['main_domain']
            
            # 네이밍 로직
            if avg_clicks > 50:
                if specialization > 0.7:
                    name = f"Heavy {main_domain} Users"
                else:
                    name = "Multi-Domain Power Users"
            elif conversion_rate > 0.6:
                name = f"High-Converting {main_domain} Users"
            elif specialization > 0.8:
                name = f"{main_domain} Specialists"
            elif avg_clicks < 10:
                name = "Light Users"
            else:
                name = f"Regular {main_domain} Users"
            
            cluster_names[cluster_id] = name
        
        return cluster_names
    
    def visualize_clusters(self, figsize: Tuple[int, int] = (20, 15)):
        """클러스터 시각화"""
        if self.clustered_data is None:
            raise ValueError("먼저 클러스터링을 수행하세요.")
        
        fig = plt.figure(figsize=figsize)
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)
        
        colors = plt.cm.Set3(np.linspace(0, 1, len(self.clustered_data['cluster'].unique())))
        
        # 1. PCA 산점도
        ax1 = fig.add_subplot(gs[0, :2])
        # 클러스터링에 사용된 피처로 PCA 수행
        X, _ = self.prepare_clustering_features()
        X_scaled = self.scaler.transform(X) if self.scaler else StandardScaler().fit_transform(X)
        
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X_scaled)
        
        for i, cluster in enumerate(sorted(self.clustered_data['cluster'].unique())):
            mask = self.clustered_data['cluster'] == cluster
            ax1.scatter(X_pca[mask, 0], X_pca[mask, 1], 
                       c=[colors[i]], label=f'Cluster {cluster}', alpha=0.6, s=30)
        
        ax1.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})')
        ax1.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})')
        
        # 매체사 정보 포함
        title = 'User Clusters Distribution (PCA)'
        if self.current_media_id:
            title += f' - Media {self.current_media_id}'
        ax1.set_title(title)
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 2. 클러스터 크기
        ax2 = fig.add_subplot(gs[0, 2])
        cluster_sizes = self.clustered_data['cluster'].value_counts().sort_index()
        ax2.pie(cluster_sizes, labels=[f'C{i}' for i in cluster_sizes.index], 
                autopct='%1.1f%%', colors=colors[:len(cluster_sizes)])
        ax2.set_title('Cluster Size Distribution')
        
        # 3. 성과 지표 히트맵
        ax3 = fig.add_subplot(gs[1, :])
        performance_metrics = ['avg_clicks', 'avg_conversions', 'avg_conv_rate', 'avg_domain_diversity']
        cluster_performance = self.cluster_summary[performance_metrics]
        
        # 정규화
        cluster_performance_norm = (cluster_performance - cluster_performance.min()) / (cluster_performance.max() - cluster_performance.min())
        
        sns.heatmap(cluster_performance_norm.T, annot=True, fmt='.2f', 
                    cmap='RdYlBu_r', ax=ax3, cbar_kws={'shrink': 0.8})
        ax3.set_title('Performance Metrics by Cluster (Normalized)')
        ax3.set_xlabel('Cluster')
        
        # 4. 전환율 vs 클릭수 산점도
        ax4 = fig.add_subplot(gs[2, :])
        scatter = ax4.scatter(self.cluster_summary['avg_clicks'], self.cluster_summary['avg_conv_rate'],
                             s=self.cluster_summary['user_count']/5, 
                             c=colors[:len(self.cluster_summary)], alpha=0.7)
        
        for i, (x, y) in enumerate(zip(self.cluster_summary['avg_clicks'], self.cluster_summary['avg_conv_rate'])):
            ax4.annotate(f'C{i}', (x, y), xytext=(5, 5), textcoords='offset points', fontweight='bold')
        
        ax4.set_xlabel('Average Clicks')
        ax4.set_ylabel('Average Conversion Rate')
        ax4.set_title('Engagement vs Performance (bubble size = user count)')
        ax4.grid(True, alpha=0.3)
        
        main_title = 'Comprehensive User Cluster Analysis'
        if self.current_media_id:
            main_title += f' - Media {self.current_media_id}'
        plt.suptitle(main_title, fontsize=16, y=0.98)
        plt.tight_layout()
        plt.show()
    
    def generate_business_insights(self) -> str:
        """비즈니스 인사이트 생성"""
        if self.clustered_data is None:
            raise ValueError("먼저 클러스터링을 수행하세요.")
        
        analysis = self.analyze_clusters()
        cluster_names = self.generate_cluster_names(analysis)
        
        media_info = f" (매체사 {self.current_media_id})" if self.current_media_id else ""
        insights = f"=== 유저 세그먼테이션 비즈니스 인사이트{media_info} ===\n\n"
        
        # 각 클러스터별 인사이트
        for cluster_id, info in analysis.items():
            name = cluster_names[cluster_id]
            insights += f"【{name} (Cluster {cluster_id})】\n"
            insights += f"• 규모: {info['size']:,}명 ({info['percentage']:.1f}%)\n"
            insights += f"• 활동량: 평균 {info['avg_clicks']:.1f}회 클릭\n"
            insights += f"• 전환율: {info['conversion_rate']:.1%}\n"
            insights += f"• 주요 관심사: {info['main_domain']}\n"
            
            # 추천 전략
            if info['conversion_rate'] > 0.5:
                insights += f"• 전략: 고전환 세그먼트, 프리미엄 상품 추천\n"
            elif info['avg_clicks'] > 50:
                insights += f"• 전략: 파워유저, VIP 프로그램 및 다양한 상품 노출\n"
            elif info['specialization'] > 0.7:
                insights += f"• 전략: 전문 세그먼트, {info['main_domain']} 중심 맞춤 추천\n"
            else:
                insights += f"• 전략: 일반 세그먼트, 관심 확장 및 재참여 유도\n"
            
            insights += "\n"
        
        # 전체 요약
        best_conversion = max(analysis.values(), key=lambda x: x['conversion_rate'])
        largest_segment = max(analysis.values(), key=lambda x: x['size'])
        
        insights += "=== 핵심 추천사항 ===\n"
        insights += f"• 최고 수익성: Cluster {[k for k, v in analysis.items() if v == best_conversion][0]} (전환율 {best_conversion['conversion_rate']:.1%})\n"
        insights += f"• 최대 규모: Cluster {[k for k, v in analysis.items() if v == largest_segment][0]} ({largest_segment['size']:,}명)\n"
        insights += f"• 총 세그먼트: {len(analysis)}개\n"
        
        return insights
    
    def export_results(self, filepath: str = None):
        """결과 내보내기"""
        if self.clustered_data is None:
            raise ValueError("먼저 클러스터링을 수행하세요.")
        
        if filepath is None:
            if self.current_media_id:
                filepath = f"user_segmentation_media_{self.current_media_id}.csv"
            else:
                filepath = "user_segmentation_results.csv"
        
        # 주요 컬럼만 선택
        export_cols = [self.user_id_col, 'cluster', 'total_clicks', 'total_conversions', 
                      'conversion_rate', 'main_domain', 'domain_diversity']
        
        self.clustered_data[export_cols].to_csv(filepath, index=False)
        print(f"결과가 {filepath}에 저장되었습니다.")
    
    def run_media_analysis(self, df: pd.DataFrame, media_id: int, 
                          max_k: int = 8, export_results: bool = True) -> Dict:
        """
        특정 매체사에 대한 세그먼테이션 분석
        
        Args:
            df: 전체 데이터
            media_id: 분석할 매체사 ID
            max_k: 최대 클러스터 수
            export_results: 결과 파일 저장 여부
        
        Returns:
            results: 분석 결과 딕셔너리
        """
        print(f"=== 매체사 {media_id} 세그먼테이션 분석 시작 ===\n")
        
        # 1. 매체사별 데이터 필터링
        media_df = self.filter_by_media(df, media_id)
        
        # 2. 전체 파이프라인 실행
        results = self.run_full_pipeline(media_df, max_k, 
                                       export_path=f"media_{media_id}_segmentation.csv" if export_results else None)
        
        # 3. 매체사별 결과 저장
        self.media_results[media_id] = {
            'results': results,
            'cluster_summary': self.cluster_summary.copy(),
            'analysis_results': self.analyze_clusters(),
            'cluster_names': self.generate_cluster_names(self.analyze_clusters())
        }
        
        return results
    
    def run_multi_media_analysis(self, df: pd.DataFrame, 
                                media_ids: List[int] = None, 
                                max_k: int = 8, 
                                min_users: int = 50) -> Dict:
        """
        여러 매체사에 대한 일괄 분석
        
        Args:
            df: 전체 데이터
            media_ids: 분석할 매체사 ID 리스트 (None이면 모든 매체사)
            max_k: 최대 클러스터 수
            min_users: 최소 유저 수 기준
        
        Returns:
            multi_results: 매체사별 분석 결과
        """
        print("=== 다중 매체사 세그먼테이션 분석 시작 ===\n")
        
        # 분석 대상 매체사 결정
        if media_ids is None:
            available_medias = self.get_available_medias(df)
            # 최소 유저 수 조건을 만족하는 매체사만 선택
            media_ids = []
            for media_id in available_medias:
                user_count = df[df[self.media_id_col] == media_id][self.user_id_col].nunique()
                if user_count >= min_users:
                    media_ids.append(media_id)
                else:
                    print(f"매체사 {media_id} 제외 (유저 수: {user_count}명 < {min_users}명)")
        
        print(f"분석 대상 매체사: {media_ids}\n")
        
        # 각 매체사별 분석 실행
        for media_id in media_ids:
            try:
                print(f"\n{'='*50}")
                self.run_media_analysis(df, media_id, max_k, export_results=True)
                print(f"매체사 {media_id} 분석 완료")
            except Exception as e:
                print(f"매체사 {media_id} 분석 실패: {str(e)}")
                continue
        
        # 결과 비교 분석
        self.compare_media_results()
        
        return self.media_results
    
    def compare_media_results(self):
        """매체사별 결과 비교 분석"""
        if not self.media_results:
            print("비교할 매체사 결과가 없습니다.")
            return
        
        print("\n=== 매체사별 세그먼테이션 비교 분석 ===")
        
        # 비교 테이블 생성
        comparison_data = []
        for media_id, data in self.media_results.items():
            summary = data['cluster_summary']
            total_users = summary['user_count'].sum()
            avg_conversion_rate = (summary['avg_conv_rate'] * summary['user_count']).sum() / total_users
            
            comparison_data.append({
                'Media_ID': media_id,
                'Total_Users': total_users,
                'Num_Clusters': len(summary),
                'Avg_Conversion_Rate': avg_conversion_rate,
                'Avg_Clicks_Per_User': (summary['avg_clicks'] * summary['user_count']).sum() / total_users,
                'Best_Segment_Conv_Rate': summary['avg_conv_rate'].max()
            })
        
        comparison_df = pd.DataFrame(comparison_data)
        print(comparison_df.round(3))
        
        # 최고 성과 매체사
        best_media = comparison_df.loc[comparison_df['Avg_Conversion_Rate'].idxmax(), 'Media_ID']
        print(f"\n최고 전환율 매체사: {best_media} ({comparison_df['Avg_Conversion_Rate'].max():.1%})")
        
        # 가장 큰 매체사
        largest_media = comparison_df.loc[comparison_df['Total_Users'].idxmax(), 'Media_ID']
        print(f"최대 규모 매체사: {largest_media} ({comparison_df['Total_Users'].max():,}명)")
    
    def run_full_pipeline(self, df: pd.DataFrame, max_k: int = 8, 
                         export_path: str = None) -> Dict:
        """전체 파이프라인 실행"""
        media_info = f" (매체사 {self.current_media_id})" if self.current_media_id else ""
        print(f"=== 유저 세그먼테이션 파이프라인 시작{media_info} ===\n")
        
        # 1. 피처 생성
        self.create_user_features(df)
        
        # 2. 클러스터링 준비
        X, feature_names = self.prepare_clustering_features()
        
        # 3. 최적 클러스터 수 탐색
        optimal_k = self.find_optimal_clusters(X, max_k)
        
        # 4. 클러스터링 수행
        clustered_data = self.perform_clustering(X, optimal_k)
        
        # 5. 분석 및 시각화
        analysis_results = self.analyze_clusters()
        cluster_names = self.generate_cluster_names(analysis_results)
        
        # 6. 결과 정리
        results = {
            'clustered_data': clustered_data,
            'cluster_summary': self.cluster_summary,
            'analysis_results': analysis_results,
            'cluster_names': cluster_names,
            'optimal_k': optimal_k,
            'feature_names': feature_names,
            'media_id': self.current_media_id
        }
        
        # 7. 인사이트 출력
        insights = self.generate_business_insights()
        print(insights)
        
        # 8. 결과 내보내기
        if export_path:
            self.export_results(export_path)
        
        print("=== 파이프라인 완료 ===")
        
        return results

# 사용 예시
"""
# 기본 사용법
pipeline = UserSegmentationPipeline()

# 1. 사용 가능한 매체사 확인
available_medias = pipeline.get_available_medias(records_cln)

# 2. 특정 매체사 하나만 분석
results_270 = pipeline.run_media_analysis(records_cln, media_id=270, max_k=8)
pipeline.visualize_clusters()

# 3. 여러 매체사 일괄 분석
multi_results = pipeline.run_multi_media_analysis(records_cln, 
                                                 media_ids=[270, 280, 290], 
                                                 max_k=8, 
                                                 min_users=100)

# 4. 전체 데이터 분석 (매체사 구분 없이)
pipeline_all = UserSegmentationPipeline()
results_all = pipeline_all.run_full_pipeline(records_cln, max_k=8, export_path="all_users_segmentation.csv")

# 5. 개별 단계 실행 (매체사별)
pipeline_manual = UserSegmentationPipeline()
mda_270 = pipeline_manual.filter_by_media(records_cln, 270)
pipeline_manual.create_user_features(mda_270)
X, features = pipeline_manual.prepare_clustering_features()
optimal_k = pipeline_manual.find_optimal_clusters(X)
clustered_data = pipeline_manual.perform_clustering(X, optimal_k)
pipeline_manual.visualize_clusters()
"""
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List, Tuple, Optional, Union
import warnings
warnings.filterwarnings('ignore')

class UserSegmentationPipeline:
    """
    광고 플랫폼 유저 세그먼테이션을 위한 종합 파이프라인 (매체사별 분석 지원)
    
    주요 기능:
    1. 매체사별 데이터 필터링
    2. 유저별 종합 피처 생성
    3. 자동 클러스터링 (최적 k 탐색)
    4. 클러스터 특성 분석 및 시각화
    5. 매체사별/통합 비즈니스 인사이트 생성
    """
    
    def __init__(self, user_id_col: str = 'dvc_idx', 
                 media_id_col: str = 'mda_idx',
                 domain_col: str = 'domain'):
        self.user_id_col = user_id_col
        self.media_id_col = media_id_col
        self.domain_col = domain_col
        
        # 초기화
        self.user_features = None
        self.clustered_data = None
        self.cluster_model = None
        self.scaler = None
        self.label_encoders = {}
        self.cluster_summary = None
        self.optimal_k = None
        
        # 매체사별 분석 결과 저장
        self.media_results = {}
        self.current_media_id = None
        
    def get_available_medias(self, df: pd.DataFrame) -> List[int]:
        """사용 가능한 매체사 목록 반환"""
        media_list = sorted(df[self.media_id_col].unique())
        media_counts = df[self.media_id_col].value_counts().sort_index()
        
        print("=== 사용 가능한 매체사 ===")
        for media_id in media_list:
            count = media_counts[media_id]
            users = df[df[self.media_id_col] == media_id][self.user_id_col].nunique()
            print(f"매체사 {media_id}: {count:,}개 레코드, {users:,}명 유저")
        
        return media_list
    
    def filter_by_media(self, df: pd.DataFrame, media_id: int, 
                       min_users: int = 50, min_records: int = 100) -> pd.DataFrame:
        """
        특정 매체사로 데이터 필터링
        
        Args:
            df: 전체 데이터
            media_id: 매체사 ID
            min_users: 최소 유저 수 (기본값: 50)
            min_records: 최소 레코드 수 (기본값: 100)
        
        Returns:
            filtered_df: 필터링된 데이터
        """
        # 매체사 필터링
        media_df = df[df[self.media_id_col] == media_id].copy()
        
        # 데이터 검증
        if len(media_df) < min_records:
            raise ValueError(f"매체사 {media_id}의 레코드가 {len(media_df)}개로 최소 기준({min_records})에 미달합니다.")
        
        unique_users = media_df[self.user_id_col].nunique()
        if unique_users < min_users:
            raise ValueError(f"매체사 {media_id}의 유저가 {unique_users}명으로 최소 기준({min_users})에 미달합니다.")
        
        print(f"매체사 {media_id} 데이터: {len(media_df):,}개 레코드, {unique_users:,}명 유저")
        self.current_media_id = media_id
        
        return media_df
    
    def create_user_features(self, df: pd.DataFrame) -> pd.DataFrame:
        """
        원본 데이터에서 유저별 종합 피처 생성
        
        Args:
            df: 광고 클릭 데이터 (click_key, ads_idx, dvc_idx, conversion, domain 등 포함)
        
        Returns:
            user_features: 유저별 피처 데이터프레임
        """
        print("유저 피처 생성 중...")
        
        # 시간 관련 피처 생성
        df['click_date'] = pd.to_datetime(df['click_date'])
        df['hour'] = df['click_date'].dt.hour
        df['is_weekend'] = df['click_date'].dt.weekday >= 5
        
        # 1. 기본 활동 피처
        basic_features = df.groupby(self.user_id_col).agg({
            'user_ip': 'nunique',
            'click_key': 'count',
            'conversion': ['sum', 'mean'],
            'contract_price': 'sum',
            'click_date': 'nunique',
            'rwd_idx': 'nunique'
        })
        
        basic_features.columns = ['ip_count', 'total_clicks', 'total_conversions', 
                                'avg_conversion_rate', 'total_revenue', 'visit_days', 'reward_diversity']
        
        # 2. 시간 패턴 피처
        time_features = df.groupby(self.user_id_col).agg({
            'hour': ['mean', lambda x: x.mode()[0] if len(x.mode()) > 0 else 0, 'nunique'],
            'is_weekend': 'mean'
        })
        time_features.columns = ['avg_hour', 'main_active_hour', 'hour_diversity', 'weekend_activity_ratio']
        
        # 3. 다양성 피처
        diversity_features = df.groupby(self.user_id_col).agg({
            'ads_idx': 'nunique',
            self.domain_col: 'nunique',
            'ads_category': 'nunique',
            'ads_type': 'nunique'
        })
        diversity_features.columns = ['ad_diversity', 'domain_diversity', 'category_diversity', 'type_diversity']
        
        # 4. 선호도 피처 (최빈값)
        preference_features = df.groupby(self.user_id_col).agg({
            self.domain_col: lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown',
            'ads_category': lambda x: x.mode()[0] if len(x.mode()) > 0 else 0,
            'ads_type': lambda x: x.mode()[0] if len(x.mode()) > 0 else 0,
            'ads_os_type': lambda x: x.mode()[0] if len(x.mode()) > 0 else 0
        })
        preference_features.columns = ['main_domain', 'main_category', 'main_type', 'main_os']
        
        # 5. 도메인 구성비
        domain_features = self._create_domain_ratios(df)
        
        # 6. 전환 관련 피처
        conversion_features = self._create_conversion_features(df)
        
        # 모든 피처 통합
        user_features = basic_features.copy()
        for feature_table in [time_features, diversity_features, preference_features, 
                             domain_features, conversion_features]:
            user_features = user_features.merge(feature_table, left_index=True, right_index=True, how='left')
        
        # 결측치 처리
        numeric_cols = user_features.select_dtypes(include=[np.number]).columns
        user_features[numeric_cols] = user_features[numeric_cols].fillna(0)
        
        categorical_cols = user_features.select_dtypes(include=['object']).columns
        user_features[categorical_cols] = user_features[categorical_cols].fillna('Unknown')
        
        # 파생 변수
        user_features['conversion_rate'] = user_features['total_conversions'] / user_features['total_clicks']
        user_features['revenue_per_click'] = user_features['total_revenue'] / user_features['total_clicks']
        user_features['clicks_per_day'] = user_features['total_clicks'] / user_features['visit_days']
        
        # 무한값 처리
        user_features = user_features.replace([np.inf, -np.inf], 0)
        
        self.user_features = user_features.reset_index()
        print(f"유저 피처 생성 완료: {self.user_features.shape}")
        
        return self.user_features
    
    def _create_domain_ratios(self, df: pd.DataFrame) -> pd.DataFrame:
        """도메인 참여 비율 계산"""
        user_domain_matrix = df.groupby([self.user_id_col, self.domain_col]).size().unstack(fill_value=0)
        user_domain_ratios = user_domain_matrix.div(user_domain_matrix.sum(axis=1), axis=0)
        
        # 상위 5개 도메인만 선택
        top_domains = user_domain_matrix.sum().nlargest(5).index
        domain_features = pd.DataFrame(index=user_domain_ratios.index)
        
        for domain in top_domains:
            if domain in user_domain_ratios.columns:
                domain_features[f'{domain}_ratio'] = user_domain_ratios[domain]
        
        # 도메인 전문성 지표
        domain_features['max_domain_ratio'] = user_domain_ratios.max(axis=1)
        domain_features['domain_entropy'] = -np.sum(
            user_domain_ratios * np.log(user_domain_ratios + 1e-10), axis=1
        )
        
        return domain_features
    
    def _create_conversion_features(self, df: pd.DataFrame) -> pd.DataFrame:
        """전환 관련 피처 생성"""
        conversion_data = df[df['conversion'] > 0]
        all_users = pd.DataFrame(index=df[self.user_id_col].unique())
        
        if len(conversion_data) > 0:
            conv_features = conversion_data.groupby(self.user_id_col).agg({
                'rwd_idx': 'mean',
                'ads_idx': 'nunique',
                self.domain_col: lambda x: x.mode()[0] if len(x.mode()) > 0 else 'No_Conversion'
            })
            conv_features.columns = ['avg_conversion_reward', 'conversion_ad_diversity', 'main_conversion_domain']
        else:
            conv_features = pd.DataFrame(columns=['avg_conversion_reward', 'conversion_ad_diversity', 'main_conversion_domain'])
        
        conversion_features = all_users.merge(conv_features, left_index=True, right_index=True, how='left')
        conversion_features = conversion_features.fillna({'avg_conversion_reward': 0, 
                                                        'conversion_ad_diversity': 0,
                                                        'main_conversion_domain': 'No_Conversion'})
        
        return conversion_features
    
    def prepare_clustering_features(self, categorical_cols: List[str] = None) -> Tuple[np.ndarray, List[str]]:
        """클러스터링용 피처 준비 및 전처리"""
        if self.user_features is None:
            raise ValueError("먼저 create_user_features()를 실행하세요.")
        
        if categorical_cols is None:
            categorical_cols = ['main_domain', 'main_conversion_domain']
        
        # 범주형 변수 인코딩
        for col in categorical_cols:
            if col in self.user_features.columns:
                le = LabelEncoder()
                self.user_features[f'{col}_encoded'] = le.fit_transform(self.user_features[col].astype(str))
                self.label_encoders[col] = le
        
        # 클러스터링용 피처 선택
        clustering_features = [
            'total_clicks', 'total_conversions', 'conversion_rate', 'domain_diversity',
            'weekend_activity_ratio', 'hour_diversity', 'clicks_per_day'
        ]
        
        # 인코딩된 범주형 피처 추가
        encoded_features = [f'{col}_encoded' for col in categorical_cols if col in self.user_features.columns]
        clustering_features.extend(encoded_features)
        
        # 도메인 비율 피처 추가
        ratio_features = [col for col in self.user_features.columns if '_ratio' in col and col != 'weekend_activity_ratio']
        clustering_features.extend(ratio_features)
        
        # 사용 가능한 피처만 선택
        available_features = [col for col in clustering_features if col in self.user_features.columns]
        
        X = self.user_features[available_features].copy()
        
        # 이상치 제거 (99 percentile 기준)
        for col in ['total_clicks', 'total_revenue']:
            if col in X.columns:
                percentile_99 = np.percentile(X[col], 99)
                X[col] = X[col].clip(upper=percentile_99)
        
        # 무한값 및 NaN 처리
        X = X.replace([np.inf, -np.inf], np.nan).fillna(0)
        
        print(f"클러스터링 피처 준비 완료: {len(available_features)}개 피처")
        
        return X.values, available_features
    
    def find_optimal_clusters(self, X: np.ndarray, max_k: int = 8) -> int:
        """최적 클러스터 수 탐색"""
        print("최적 클러스터 수 탐색 중...")
        
        # 표준화
        self.scaler = StandardScaler()
        X_scaled = self.scaler.fit_transform(X)
        
        inertias = []
        silhouette_scores = []
        K_range = range(2, max_k + 1)
        
        for k in K_range:
            kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
            cluster_labels = kmeans.fit_predict(X_scaled)
            
            inertias.append(kmeans.inertia_)
            silhouette_avg = silhouette_score(X_scaled, cluster_labels)
            silhouette_scores.append(silhouette_avg)
            
            print(f"k={k}: Silhouette Score={silhouette_avg:.3f}")
        
        self.optimal_k = K_range[np.argmax(silhouette_scores)]
        print(f"최적 클러스터 수: {self.optimal_k}")
        
        return self.optimal_k
    
    def perform_clustering(self, X: np.ndarray, n_clusters: int = None) -> pd.DataFrame:
        """클러스터링 수행"""
        if n_clusters is None:
            n_clusters = self.optimal_k
        
        if self.scaler is None:
            self.scaler = StandardScaler()
            X_scaled = self.scaler.fit_transform(X)
        else:
            X_scaled = self.scaler.transform(X)
        
        print(f"{n_clusters}개 클러스터로 클러스터링 수행 중...")
        
        self.cluster_model = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        cluster_labels = self.cluster_model.fit_predict(X_scaled)
        
        # 클러스터 라벨 추가
        self.user_features['cluster'] = cluster_labels
        
        # 클러스터별 특성 분석
        self.cluster_summary = self.user_features.groupby('cluster').agg({
            self.user_id_col: 'count',
            'total_clicks': 'mean',
            'total_conversions': 'mean',
            'conversion_rate': 'mean',
            'domain_diversity': 'mean',
            'max_domain_ratio': 'mean'
        }).round(2)
        
        self.cluster_summary.columns = ['user_count', 'avg_clicks', 'avg_conversions', 
                                      'avg_conv_rate', 'avg_domain_diversity', 'avg_specialization']
        
        self.clustered_data = self.user_features.copy()
        
        print("클러스터링 완료!")
        print(self.cluster_summary)
        
        return self.clustered_data
    
    def analyze_clusters(self) -> Dict:
        """클러스터별 상세 분석"""
        if self.clustered_data is None:
            raise ValueError("먼저 클러스터링을 수행하세요.")
        
        analysis_results = {}
        
        for cluster_id in sorted(self.clustered_data['cluster'].unique()):
            cluster_data = self.clustered_data[self.clustered_data['cluster'] == cluster_id]
            
            # 도메인 선호도
            domain_prefs = {}
            ratio_cols = [col for col in cluster_data.columns if '_ratio' in col and col != 'weekend_activity_ratio']
            for col in ratio_cols:
                avg_ratio = cluster_data[col].mean()
                if avg_ratio > 0.05:  # 5% 이상인 것만
                    domain_prefs[col.replace('_ratio', '')] = avg_ratio
            
            analysis_results[cluster_id] = {
                'size': len(cluster_data),
                'percentage': len(cluster_data) / len(self.clustered_data) * 100,
                'avg_clicks': cluster_data['total_clicks'].mean(),
                'avg_conversions': cluster_data['total_conversions'].mean(),
                'conversion_rate': cluster_data['conversion_rate'].mean(),
                'domain_diversity': cluster_data['domain_diversity'].mean(),
                'specialization': cluster_data['max_domain_ratio'].mean(),
                'weekend_activity': cluster_data['weekend_activity_ratio'].mean(),
                'domain_preferences': domain_prefs,
                'main_domain': cluster_data['main_domain'].mode()[0] if len(cluster_data['main_domain'].mode()) > 0 else 'Mixed'
            }
        
        return analysis_results
    
    def generate_cluster_names(self, analysis_results: Dict) -> Dict[int, str]:
        """클러스터별 비즈니스 네이밍"""
        cluster_names = {}
        
        for cluster_id, info in analysis_results.items():
            avg_clicks = info['avg_clicks']
            conversion_rate = info['conversion_rate']
            specialization = info['specialization']
            main_domain = info['main_domain']
            
            # 네이밍 로직
            if avg_clicks > 50:
                if specialization > 0.7:
                    name = f"Heavy {main_domain} Users"
                else:
                    name = "Multi-Domain Power Users"
            elif conversion_rate > 0.6:
                name = f"High-Converting {main_domain} Users"
            elif specialization > 0.8:
                name = f"{main_domain} Specialists"
            elif avg_clicks < 10:
                name = "Light Users"
            else:
                name = f"Regular {main_domain} Users"
            
            cluster_names[cluster_id] = name
        
        return cluster_names
    
    def visualize_clusters(self, figsize: Tuple[int, int] = (20, 15)):
        """클러스터 시각화"""
        if self.clustered_data is None:
            raise ValueError("먼저 클러스터링을 수행하세요.")
        
        fig = plt.figure(figsize=figsize)
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)
        
        colors = plt.cm.Set3(np.linspace(0, 1, len(self.clustered_data['cluster'].unique())))
        
        # 1. PCA 산점도
        ax1 = fig.add_subplot(gs[0, :2])
        # 클러스터링에 사용된 피처로 PCA 수행
        X, _ = self.prepare_clustering_features()
        X_scaled = self.scaler.transform(X) if self.scaler else StandardScaler().fit_transform(X)
        
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X_scaled)
        
        for i, cluster in enumerate(sorted(self.clustered_data['cluster'].unique())):
            mask = self.clustered_data['cluster'] == cluster
            ax1.scatter(X_pca[mask, 0], X_pca[mask, 1], 
                       c=[colors[i]], label=f'Cluster {cluster}', alpha=0.6, s=30)
        
        ax1.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})')
        ax1.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})')
        
        # 매체사 정보 포함
        title = 'User Clusters Distribution (PCA)'
        if self.current_media_id:
            title += f' - Media {self.current_media_id}'
        ax1.set_title(title)
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 2. 클러스터 크기
        ax2 = fig.add_subplot(gs[0, 2])
        cluster_sizes = self.clustered_data['cluster'].value_counts().sort_index()
        ax2.pie(cluster_sizes, labels=[f'C{i}' for i in cluster_sizes.index], 
                autopct='%1.1f%%', colors=colors[:len(cluster_sizes)])
        ax2.set_title('Cluster Size Distribution')
        
        # 3. 성과 지표 히트맵
        ax3 = fig.add_subplot(gs[1, :])
        performance_metrics = ['avg_clicks', 'avg_conversions', 'avg_conv_rate', 'avg_domain_diversity']
        cluster_performance = self.cluster_summary[performance_metrics]
        
        # 정규화
        cluster_performance_norm = (cluster_performance - cluster_performance.min()) / (cluster_performance.max() - cluster_performance.min())
        
        sns.heatmap(cluster_performance_norm.T, annot=True, fmt='.2f', 
                    cmap='RdYlBu_r', ax=ax3, cbar_kws={'shrink': 0.8})
        ax3.set_title('Performance Metrics by Cluster (Normalized)')
        ax3.set_xlabel('Cluster')
        
        # 4. 전환율 vs 클릭수 산점도
        ax4 = fig.add_subplot(gs[2, :])
        scatter = ax4.scatter(self.cluster_summary['avg_clicks'], self.cluster_summary['avg_conv_rate'],
                             s=self.cluster_summary['user_count']/5, 
                             c=colors[:len(self.cluster_summary)], alpha=0.7)
        
        for i, (x, y) in enumerate(zip(self.cluster_summary['avg_clicks'], self.cluster_summary['avg_conv_rate'])):
            ax4.annotate(f'C{i}', (x, y), xytext=(5, 5), textcoords='offset points', fontweight='bold')
        
        ax4.set_xlabel('Average Clicks')
        ax4.set_ylabel('Average Conversion Rate')
        ax4.set_title('Engagement vs Performance (bubble size = user count)')
        ax4.grid(True, alpha=0.3)
        
        main_title = 'Comprehensive User Cluster Analysis'
        if self.current_media_id:
            main_title += f' - Media {self.current_media_id}'
        plt.suptitle(main_title, fontsize=16, y=0.98)
        plt.tight_layout()
        plt.show()
    
    def generate_business_insights(self) -> str:
        """비즈니스 인사이트 생성"""
        if self.clustered_data is None:
            raise ValueError("먼저 클러스터링을 수행하세요.")
        
        analysis = self.analyze_clusters()
        cluster_names = self.generate_cluster_names(analysis)
        
        media_info = f" (매체사 {self.current_media_id})" if self.current_media_id else ""
        insights = f"=== 유저 세그먼테이션 비즈니스 인사이트{media_info} ===\n\n"
        
        # 각 클러스터별 인사이트
        for cluster_id, info in analysis.items():
            name = cluster_names[cluster_id]
            insights += f"【{name} (Cluster {cluster_id})】\n"
            insights += f"• 규모: {info['size']:,}명 ({info['percentage']:.1f}%)\n"
            insights += f"• 활동량: 평균 {info['avg_clicks']:.1f}회 클릭\n"
            insights += f"• 전환율: {info['conversion_rate']:.1%}\n"
            insights += f"• 주요 관심사: {info['main_domain']}\n"
            
            # 추천 전략
            if info['conversion_rate'] > 0.5:
                insights += f"• 전략: 고전환 세그먼트, 프리미엄 상품 추천\n"
            elif info['avg_clicks'] > 50:
                insights += f"• 전략: 파워유저, VIP 프로그램 및 다양한 상품 노출\n"
            elif info['specialization'] > 0.7:
                insights += f"• 전략: 전문 세그먼트, {info['main_domain']} 중심 맞춤 추천\n"
            else:
                insights += f"• 전략: 일반 세그먼트, 관심 확장 및 재참여 유도\n"
            
            insights += "\n"
        
        # 전체 요약
        best_conversion = max(analysis.values(), key=lambda x: x['conversion_rate'])
        largest_segment = max(analysis.values(), key=lambda x: x['size'])
        
        insights += "=== 핵심 추천사항 ===\n"
        insights += f"• 최고 수익성: Cluster {[k for k, v in analysis.items() if v == best_conversion][0]} (전환율 {best_conversion['conversion_rate']:.1%})\n"
        insights += f"• 최대 규모: Cluster {[k for k, v in analysis.items() if v == largest_segment][0]} ({largest_segment['size']:,}명)\n"
        insights += f"• 총 세그먼트: {len(analysis)}개\n"
        
        return insights
    
    def export_results(self, filepath: str = None):
        """결과 내보내기"""
        if self.clustered_data is None:
            raise ValueError("먼저 클러스터링을 수행하세요.")
        
        if filepath is None:
            if self.current_media_id:
                filepath = f"user_segmentation_media_{self.current_media_id}.csv"
            else:
                filepath = "user_segmentation_results.csv"
        
        # 주요 컬럼만 선택
        export_cols = [self.user_id_col, 'cluster', 'total_clicks', 'total_conversions', 
                      'conversion_rate', 'main_domain', 'domain_diversity']
        
        self.clustered_data[export_cols].to_csv(filepath, index=False)
        print(f"결과가 {filepath}에 저장되었습니다.")
    
    def run_media_analysis(self, df: pd.DataFrame, media_id: int, 
                          max_k: int = 8, export_results: bool = True) -> Dict:
        """
        특정 매체사에 대한 세그먼테이션 분석
        
        Args:
            df: 전체 데이터
            media_id: 분석할 매체사 ID
            max_k: 최대 클러스터 수
            export_results: 결과 파일 저장 여부
        
        Returns:
            results: 분석 결과 딕셔너리
        """
        print(f"=== 매체사 {media_id} 세그먼테이션 분석 시작 ===\n")
        
        # 1. 매체사별 데이터 필터링
        media_df = self.filter_by_media(df, media_id)
        
        # 2. 전체 파이프라인 실행
        results = self.run_full_pipeline(media_df, max_k, 
                                       export_path=f"media_{media_id}_segmentation.csv" if export_results else None)
        
        # 3. 매체사별 결과 저장
        self.media_results[media_id] = {
            'results': results,
            'cluster_summary': self.cluster_summary.copy(),
            'analysis_results': self.analyze_clusters(),
            'cluster_names': self.generate_cluster_names(self.analyze_clusters())
        }
        
        return results
    
    def run_multi_media_analysis(self, df: pd.DataFrame, 
                                media_ids: List[int] = None, 
                                max_k: int = 8, 
                                min_users: int = 50) -> Dict:
        """
        여러 매체사에 대한 일괄 분석
        
        Args:
            df: 전체 데이터
            media_ids: 분석할 매체사 ID 리스트 (None이면 모든 매체사)
            max_k: 최대 클러스터 수
            min_users: 최소 유저 수 기준
        
        Returns:
            multi_results: 매체사별 분석 결과
        """
        print("=== 다중 매체사 세그먼테이션 분석 시작 ===\n")
        
        # 분석 대상 매체사 결정
        if media_ids is None:
            available_medias = self.get_available_medias(df)
            # 최소 유저 수 조건을 만족하는 매체사만 선택
            media_ids = []
            for media_id in available_medias:
                user_count = df[df[self.media_id_col] == media_id][self.user_id_col].nunique()
                if user_count >= min_users:
                    media_ids.append(media_id)
                else:
                    print(f"매체사 {media_id} 제외 (유저 수: {user_count}명 < {min_users}명)")
        
        print(f"분석 대상 매체사: {media_ids}\n")
        
        # 각 매체사별 분석 실행
        for media_id in media_ids:
            try:
                print(f"\n{'='*50}")
                self.run_media_analysis(df, media_id, max_k, export_results=True)
                print(f"매체사 {media_id} 분석 완료")
            except Exception as e:
                print(f"매체사 {media_id} 분석 실패: {str(e)}")
                continue
        
        # 결과 비교 분석
        self.compare_media_results()
        
        return self.media_results
    
    def compare_media_results(self):
        """매체사별 결과 비교 분석"""
        if not self.media_results:
            print("비교할 매체사 결과가 없습니다.")
            return
        
        print("\n=== 매체사별 세그먼테이션 비교 분석 ===")
        
        # 비교 테이블 생성
        comparison_data = []
        for media_id, data in self.media_results.items():
            summary = data['cluster_summary']
            total_users = summary['user_count'].sum()
            avg_conversion_rate = (summary['avg_conv_rate'] * summary['user_count']).sum() / total_users
            
            comparison_data.append({
                'Media_ID': media_id,
                'Total_Users': total_users,
                'Num_Clusters': len(summary),
                'Avg_Conversion_Rate': avg_conversion_rate,
                'Avg_Clicks_Per_User': (summary['avg_clicks'] * summary['user_count']).sum() / total_users,
                'Best_Segment_Conv_Rate': summary['avg_conv_rate'].max()
            })
        
        comparison_df = pd.DataFrame(comparison_data)
        print(comparison_df.round(3))
        
        # 최고 성과 매체사
        best_media = comparison_df.loc[comparison_df['Avg_Conversion_Rate'].idxmax(), 'Media_ID']
        print(f"\n최고 전환율 매체사: {best_media} ({comparison_df['Avg_Conversion_Rate'].max():.1%})")
        
        # 가장 큰 매체사
        largest_media = comparison_df.loc[comparison_df['Total_Users'].idxmax(), 'Media_ID']
        print(f"최대 규모 매체사: {largest_media} ({comparison_df['Total_Users'].max():,}명)")
    
    def run_full_pipeline(self, df: pd.DataFrame, max_k: int = 8, 
                         export_path: str = None) -> Dict:
        """전체 파이프라인 실행"""
        media_info = f" (매체사 {self.current_media_id})" if self.current_media_id else ""
        print(f"=== 유저 세그먼테이션 파이프라인 시작{media_info} ===\n")
        
        # 1. 피처 생성
        self.create_user_features(df)
        
        # 2. 클러스터링 준비
        X, feature_names = self.prepare_clustering_features()
        
        # 3. 최적 클러스터 수 탐색
        optimal_k = self.find_optimal_clusters(X, max_k)
        
        # 4. 클러스터링 수행
        clustered_data = self.perform_clustering(X, optimal_k)
        
        # 5. 분석 및 시각화
        analysis_results = self.analyze_clusters()
        cluster_names = self.generate_cluster_names(analysis_results)
        
        # 6. 결과 정리
        results = {
            'clustered_data': clustered_data,
            'cluster_summary': self.cluster_summary,
            'analysis_results': analysis_results,
            'cluster_names': cluster_names,
            'optimal_k': optimal_k,
            'feature_names': feature_names,
            'media_id': self.current_media_id
        }
        
        # 7. 인사이트 출력
        insights = self.generate_business_insights()
        print(insights)
        
        # 8. 결과 내보내기
        if export_path:
            self.export_results(export_path)
        
        print("=== 파이프라인 완료 ===")
        
        return results

# 사용 예시
"""
# 기본 사용법
pipeline = UserSegmentationPipeline()

# 1. 사용 가능한 매체사 확인
available_medias = pipeline.get_available_medias(records_cln)

# 2. 특정 매체사 하나만 분석 (종합 리포트 포함)
results_270 = pipeline.run_media_analysis(records_cln, media_id=270, max_k=8, 
                                         show_comprehensive_report=True)
pipeline.visualize_clusters()

# 3. 여러 매체사 일괄 분석 (개별 리포트 + 비교 리포트)
multi_results = pipeline.run_multi_media_analysis(records_cln, 
                                                 media_ids=[270, 280, 290], 
                                                 max_k=8, 
                                                 min_users=100,
                                                 show_individual_reports=True)

# 4. 전체 데이터 분석 (매체사 구분 없이) + 종합 리포트
pipeline_all = UserSegmentationPipeline()
results_all = pipeline_all.run_full_pipeline(records_cln, max_k=8, 
                                            export_path="all_users_segmentation.csv",
                                            show_comprehensive_report=True)

# 5. 종합 리포트만 따로 생성하기
pipeline = UserSegmentationPipeline()
results = pipeline.run_media_analysis(records_cln, media_id=270, show_comprehensive_report=False)

# 이후에 종합 리포트 생성
comprehensive_report = pipeline.generate_comprehensive_report()
print(comprehensive_report)

# 리포트를 파일로 저장
with open('detailed_analysis_report.txt', 'w', encoding='utf-8') as f:
    f.write(comprehensive_report)

# 6. 매체사 비교 분석만 원할 때
pipeline = UserSegmentationPipeline()

# 여러 매체사 분석 실행 (리포트 출력 없이)
for media_id in [270, 280, 290]:
    pipeline.run_media_analysis(records_cln, media_id, show_comprehensive_report=False)

# 비교 리포트만 생성
comparison_report = pipeline.generate_media_comparison_report()
print(comparison_report)

# 7. 개별 단계 실행 + 맞춤 해석
pipeline_manual = UserSegmentationPipeline()
mda_270 = pipeline_manual.filter_by_media(records_cln, 270)
pipeline_manual.create_user_features(mda_270)
X, features = pipeline_manual.prepare_clustering_features()
optimal_k = pipeline_manual.find_optimal_clusters(X)
clustered_data = pipeline_manual.perform_clustering(X, optimal_k)

# 각종 해석 결과 얻기
basic_insights = pipeline_manual.generate_business_insights()
comprehensive_report = pipeline_manual.generate_comprehensive_report()
analysis_results = pipeline_manual.analyze_clusters()
cluster_names = pipeline_manual.generate_cluster_names(analysis_results)

print("=== 기본 인사이트 ===")
print(basic_insights)
print("\n=== 종합 리포트 ===")
print(comprehensive_report)

# 시각화
pipeline_manual.visualize_clusters()

# 결과 저장 (데이터 + 리포트)
export_result = pipeline_manual.export_results("manual_analysis.csv", include_report=True)
print(f"저장된 파일들: {export_result}")

# 8. 특정 세그먼트에 대한 상세 분석
pipeline = UserSegmentationPipeline()
results = pipeline.run_media_analysis(records_cln, media_id=270)

# 특정 클러스터 유저들만 추출
cluster_0_users = pipeline.clustered_data[pipeline.clustered_data['cluster'] == 0]
print(f"Cluster 0 유저 수: {len(cluster_0_users)}")
print(f"Cluster 0 특성: {pipeline.analyze_clusters()[0]}")

# 9. ROI 기반 우선순위 분석
analysis = pipeline.analyze_clusters()
cluster_names = pipeline.generate_cluster_names(analysis)

# 각 세그먼트의 ROI 점수 확인
for cluster_id, info in analysis.items():
    name = cluster_names[cluster_id]
    roi_score = info['conversion_rate'] * (info['size'] / sum(i['size'] for i in analysis.values()))
    print(f"{name}: ROI 점수 {roi_score:.3f}")
"""
```
