```python
# 극값들을 고려한 세분화
def advanced_classification(df):
    def classify_advanced(row):
        price = row['contract_price']
        media = row['media_count']
        clicks = row['total_clicks'] 
        conversions = row['total_conversions']
        
        # MEGA: 극상위 (최대값의 10% 이상)
        if (price >= 145600 or    # 최대값의 10%
            media >= 10 or        # 매체사 10개 이상
            clicks >= 270000):    # 클릭 27만 이상
            return 'MEGA'
            
        # LARGE: 상위 5%
        elif (price >= df['contract_price'].quantile(0.95) or
              media >= df['media_count'].quantile(0.95) or  
              clicks >= df['total_clicks'].quantile(0.95)):
            return 'LARGE'
            
        # MEDIUM: 상위 25%
        elif (price >= df['contract_price'].quantile(0.75) or
              media >= 3 or
              clicks >= df['total_clicks'].quantile(0.75)):
            return 'MEDIUM'
            
        # SMALL: 평균 이상
        elif (clicks >= df['total_clicks'].median() and 
              conversions >= df['total_conversions'].median()):
            return 'SMALL'
            
        # MICRO: 나머지
        else:
            return 'MICRO'
    
    df['ads_level_detailed'] = df.apply(classify_advanced, axis=1)
    return df

ads_seg = advanced_classification(ads_seg)
print("\n세분화된 레벨 분포:")
print(ads_seg['ads_level_detailed'].value_counts())
```

```python
# 일별 기준으로 정규화된 지표들 계산
ads_seg['daily_clicks'] = ads_seg['total_clicks'] / ads_seg['days_active']
ads_seg['daily_users'] = ads_seg['user_count'] / ads_seg['days_active']

def classify_by_percentiles_fixed_criteria(df):
    # 규모 지표들 (고정)
    media_q90 = df['media_count'].quantile(0.90)
    media_q75 = df['media_count'].quantile(0.75)
    daily_clicks_q90 = df['daily_clicks'].quantile(0.90)
    daily_clicks_q75 = df['daily_clicks'].quantile(0.75)
    daily_users_q90 = df['daily_users'].quantile(0.90)
    daily_users_q75 = df['daily_users'].quantile(0.75)
    
    # 성과 지표들 (고정)
    daily_conv_q90 = df['daily_avg_conversions'].quantile(0.90)
    daily_conv_q75 = df['daily_avg_conversions'].quantile(0.75)
    cvr_q90 = df['cvr'].quantile(0.90)
    cvr_q75 = df['cvr'].quantile(0.75)
    
    def get_scale_level(row):
        # 매체사 수, 일별 클릭수, 일별 사용자 수 - 이 3개 고정
        media_score = (row['media_count'] >= media_q90) * 2 + (row['media_count'] >= media_q75) * 1
        clicks_score = (row['daily_clicks'] >= daily_clicks_q90) * 2 + (row['daily_clicks'] >= daily_clicks_q75) * 1
        users_score = (row['daily_users'] >= daily_users_q90) * 2 + (row['daily_users'] >= daily_users_q75) * 1
        
        total_score = media_score + clicks_score + users_score
        
        if total_score >= 4:
            return 'MEGA'
        elif total_score >= 2:
            return 'LARGE'
        elif total_score >= 1:
            return 'MEDIUM'
        else:
            return 'SMALL'
    
    def get_performance_level(row):
        # 일별 전환수, 전환율 - 이 2개 고정
        conv_score = (row['daily_avg_conversions'] >= daily_conv_q90) * 2 + (row['daily_avg_conversions'] >= daily_conv_q75) * 1
        cvr_score = (row['cvr'] >= cvr_q90) * 2 + (row['cvr'] >= cvr_q75) * 1
        
        total_score = conv_score + cvr_score
        
        if total_score >= 3:
            return 'HIGH'
        elif total_score >= 1:
            return 'GOOD'
        else:
            return 'LOW'
    
    df['scale_level'] = df.apply(get_scale_level, axis=1)
    df['performance_level'] = df.apply(get_performance_level, axis=1)
    df['ads_level'] = df['scale_level'] + '_' + df['performance_level']
    
    return df

ads_seg = classify_by_percentiles_fixed_criteria(ads_seg)

# 결과 확인
print("분위수 기반 분류 결과:")
print(ads_seg['ads_level'].value_counts())
```

```python
def classify_by_percentiles(df):
    # daily_clicks 계산
    df['daily_clicks'] = df['total_clicks'] / df['days_active']
    
    # 각 지표별 백분위 기준 점수
    def get_percentile_score(value, series):
        if value >= series.quantile(0.995): return 4  # 상위 0.5%
        elif value >= series.quantile(0.95): return 3  # 상위 5%
        elif value >= series.quantile(0.75): return 2  # 상위 25%
        elif value >= series.quantile(0.50): return 1  # 상위 50%
        else: return 0
    
    # 각 지표별 점수 계산
    df['media_score'] = df['media_count'].apply(lambda x: get_percentile_score(x, df['media_count']))
    df['conv_score'] = df['daily_avg_conversions'].apply(lambda x: get_percentile_score(x, df['daily_avg_conversions']))
    df['clicks_score'] = df['daily_clicks'].apply(lambda x: get_percentile_score(x, df['daily_clicks']))
    df['stability_score'] = df['days_active'].apply(lambda x: get_percentile_score(x, df['days_active']))
    df['cvr_score'] = df['cvr'].apply(lambda x: get_percentile_score(x, df['cvr']))
    
    # 총점 계산 (최대 20점)
    df['total_score'] = (df['media_score'] + df['conv_score'] + df['clicks_score'] + 
                        df['stability_score'] + df['cvr_score'])
    
    # 백분위 기준 분류
    def get_size_by_percentile(score, score_series):
        if score >= score_series.quantile(0.995): return 'MEGA'    # 상위 0.5%
        elif score >= score_series.quantile(0.90): return 'LARGE'  # 상위 10%
        elif score >= score_series.quantile(0.70): return 'MEDIUM' # 상위 30%
        else: return 'SMALL'
    
    df['ads_size'] = df['total_score'].apply(lambda x: get_size_by_percentile(x, df['total_score']))
    
    return df

ads_seg = classify_by_percentiles(ads_seg)

print("백분위 기준 분류:")
print(ads_seg['ads_size'].value_counts())
print(f"\n분류 비율:")
print((ads_seg['ads_size'].value_counts() / len(ads_seg) * 100).round(1))
```

```python
def standard_tier_classification(df):
    # daily_clicks 계산
    df['daily_clicks'] = df['total_clicks'] / df['days_active']
    
    # 점수 계산 (동일)
    def get_percentile_score(value, series):
        if value >= series.quantile(0.99): return 4
        elif value >= series.quantile(0.90): return 3  
        elif value >= series.quantile(0.70): return 2
        elif value >= series.quantile(0.40): return 1
        else: return 0
    
    df['media_score'] = df['media_count'].apply(lambda x: get_percentile_score(x, df['media_count']))
    df['conv_score'] = df['daily_avg_conversions'].apply(lambda x: get_percentile_score(x, df['daily_avg_conversions']))
    df['clicks_score'] = df['daily_clicks'].apply(lambda x: get_percentile_score(x, df['daily_clicks']))
    df['stability_score'] = df['days_active'].apply(lambda x: get_percentile_score(x, df['days_active']))
    df['cvr_score'] = df['cvr'].apply(lambda x: get_percentile_score(x, df['cvr']))
    
    df['total_score'] = (df['media_score'] + df['conv_score'] + df['clicks_score'] + 
                        df['stability_score'] + df['cvr_score'])
    
    # 표준 분류 (피라미드 형태)
    def get_standard_tier(score, score_series):
        if score >= score_series.quantile(0.95): return 'MEGA'    # 상위 5%
        elif score >= score_series.quantile(0.80): return 'LARGE'  # 상위 20%
        elif score >= score_series.quantile(0.30): return 'MEDIUM' # 상위 70% (가장 많음)
        else: return 'SMALL'                                      # 하위 30%
    
    df['ads_size'] = df['total_score'].apply(lambda x: get_standard_tier(x, df['total_score']))
    
    return df


ads_seg = standard_tier_classification(ads_seg)

print("백분위 기준 분류:")
print(ads_seg['ads_size'].value_counts())
print(f"\n분류 비율:")
print((ads_seg['ads_size'].value_counts() / len(ads_seg) * 100).round(1))
print(ads_seg.head())
```

```python
# 1. 매체사-광고 관계에 광고 속성 정보 추가
media_ads_with_attributes = media_ads_mapping.merge(
    ads_list[['ads_idx', 'ads_category', 'ads_os_type', 'domain']], 
    on='ads_idx', 
    how='left'
)

# 2. 각 속성별로 매체사별 분포 계산
def calculate_attribute_distribution(df, attribute_col):
    # 매체사별 해당 속성 분포
    attr_dist = df.groupby(['mda_idx', attribute_col]).size().unstack(fill_value=0)
    
    # 총 광고 수로 나누어 비율 계산
    attr_dist_pct = attr_dist.div(attr_dist.sum(axis=1), axis=0) * 100
    
    # 컬럼명에 prefix 추가
    attr_dist_pct.columns = [f'{attribute_col}_{col}_pct' for col in attr_dist_pct.columns]
    
    return attr_dist_pct

# 3. 각 속성별 비율 계산
category_dist = calculate_attribute_distribution(media_ads_with_attributes, 'ads_category')
os_dist = calculate_attribute_distribution(media_ads_with_attributes, 'ads_os_type')  
domain_dist = calculate_attribute_distribution(media_ads_with_attributes, 'domain')

# 4. 모든 분포 정보를 기존 테이블에 조인
media_portfolio = media_portfolio.merge(
    category_dist, on='mda_idx', how='left'
).merge(
    os_dist, on='mda_idx', how='left'
).merge(
    domain_dist, on='mda_idx', how='left'
)

# 5. 결과 확인
print(f"최종 테이블 크기: {media_portfolio.shape}")
print("추가된 컬럼들:")
new_cols = [col for col in media_portfolio.columns if '_pct' in col]
print(new_cols[:10])  # 처음 10개만 출력

# 샘플 확인
print("\n매체사별 속성 분포 (상위 5개 매체사):")
sample_cols = ['mda_idx', 'total_ads'] + new_cols[:6]  # 샘플로 몇 개만
print(media_portfolio[sample_cols].head())
```

```python
# 1. click 데이터에 광고 단가 정보 조인
click_with_price = click.merge(
    ads_seg[['ads_idx', 'contract_price', 'media_price']], 
    on='ads_idx', 
    how='left'
)

# 2. 매체사-광고별 수익 계산
click_with_price['profit_per_conversion'] = (
    click_with_price['contract_price'] - click_with_price['media_price']
)

# 3. 매체사-광고별 총 수익 계산
media_ads_profit = click_with_price.groupby(['mda_idx', 'ads_idx']).agg({
    'conversion': 'sum',
    'profit_per_conversion': 'first'  # 광고별로 동일한 값
}).reset_index()

media_ads_profit['total_profit'] = (
    media_ads_profit['conversion'] * media_ads_profit['profit_per_conversion']
)

# 4. 매체사별 총 예상 수익 집계
media_expected_profit = media_ads_profit.groupby('mda_idx').agg({
    'total_profit': 'sum'
}).reset_index()

media_expected_profit.columns = ['mda_idx', 'expected_total_profit']

# 5. media_portfolio에 조인
media_portfolio = media_portfolio.merge(
    media_expected_profit, 
    on='mda_idx', 
    how='left'
)

# NaN 처리
media_portfolio['expected_total_profit'] = (
    media_portfolio['expected_total_profit'].fillna(0)
)

# 결과 확인
print("매체사별 예상 수익 상위 10개:")
print(media_portfolio.sort_values('expected_total_profit', ascending=False)
      [['mda_idx', 'total_ads', 'total_conversions', 'expected_total_profit']].head(10))
```
